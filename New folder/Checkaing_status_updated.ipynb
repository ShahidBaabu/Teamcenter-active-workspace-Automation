```python
from selenium import webdriver
from selenium.webdriver.edge.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from datetime import date
import pandas as pd
import csv
import time

#Initializing Teamcenter active workspace on our webdriver
service = Service(executable_path= './chromedriver.exe')
driver = webdriver.Chrome(service=service)
driver.get('https://pdment01p:3000/')
wait = WebDriverWait(driver, 0.5)

data = pd.read_csv('Doc_numbers.csv')
username = data.iloc[0,1]
password = data.iloc[0,2]

def trying_fnctn(xpath):
    element = ''
    try:
        wait.until(EC.visibility_of_element_located((By.XPATH, xpath)))
    except:
        trying_fnctn(xpath)
        
    element = wait.until(EC.visibility_of_element_located((By.XPATH, xpath)))
    return element

#Entering username and password
xpath = '/html/body/div/form/div/div/div[2]/div/ul/li[1]/input'
input_username = trying_fnctn(xpath)
input_username.send_keys(username)
xpath = '//input[@ng-model = "password"]'
input_password = trying_fnctn(xpath)
input_password.send_keys(password)

xpath = '//button[@class = "aw-base-button ng-binding"]'
signin_button = trying_fnctn(xpath)
signin_button.click()
time.sleep(5)


#data.columns = ['Item-ID']
items = data['Item_Id'].tolist()
result_list = []
version_list = []

for item in items:
    #Operating search
    #search = wait.until(EC.visibility_of_element_located((By.XPATH, '//*[@id="cmdSearch"]')))
    #search.click()
    #time.sleep(5)
    
    xpath = '//*[@id="main-view"]/aw-include/div/div/div/div/div/ui-view/aw-page/div/div/div/div[1]/aw-header/div/div/div[2]/aw-global-search/div/aw-include/div/aw-search-global/div/div[1]/div[1]/div[2]/div[2]/div[2]/aw-link/div/a'
    Advanced_btn = trying_fnctn(xpath)
    Advanced_btn.click()
    time.sleep(3)

    # Opening the item and sorting
    xpath = '//*[@id="aw_navigation"]/div/aw-include/div/div/div[2]/div/form/aw-tab-set/div/form[1]/div/form/aw-widget/div/div/div/aw-property-val/div/div/aw-property-string-val/div/div/aw-property-text-area-val/div/aw-property-error/div/textarea'
    itemID_Input = trying_fnctn(xpath)
    itemID_Input.clear()
    itemID_Input.send_keys(item)
    time.sleep(3)
    
    xpath = '//*[@id="aw_navigation"]/div/aw-include/div/div/div[2]/div/div/button[2]'
    search_button = trying_fnctn(xpath)
    search_button.click()
    time.sleep(3)
    
    xpath = '//*[@id="Awp0ShowObjectCell"]'
    open_option = trying_fnctn(xpath)
    open_option.click()
    time.sleep(3)
    
    xpath = '//*[@id="ObjectSet_1_Provider"]/div[2]/div[3]/div[1]/div[1]/div[2]/div'
    revision = trying_fnctn(xpath)
    revision.click()
    time.sleep(3)
    
    xpath = '//*[@id="ObjectSet_1_Provider_menu"]/li[2]/i'
    sort_descending = trying_fnctn(xpath)
    sort_descending.click()
    time.sleep(3)

    #Finding the state
    xpath = '//*[@id="ObjectSet_1_Provider"]/div[2]/div[3]/div[2]/div/div[1]/div[2]/div'
    div_elem = trying_fnctn(xpath)

    img_elem = div_elem.find_elements(By.TAG_NAME, 'img')
    time.sleep(3)
    

    if len(img_elem) > 0:
        sol = (img_elem[0].get_attribute('title'))
    else:
        sol = ('In Work')
        
    result_list.append(sol)
    
    #Finding the version
    xpath = '/html/body/div[1]/aw-include/div/div/div/div/div/ui-view/aw-showobject-page/div/div/div/div[3]/div/ng-transclude/aw-xrt-sublocation/aw-sublocation/div/div[2]/div/aw-sublocation-body/div/aw-xrt-2/aw-walker-view/div[1]/aw-walker-element/div[2]/form/aw-walker-element/div/div/aw-walker-objectset/div[2]/aw-splm-table/div/div[2]/div[3]/div[2]/div/div/div[1]/div/div'
    version_elem = trying_fnctn(xpath)
    time.sleep(3)
    
    version_list.append(str(version_elem.text))
    
    xpath = '//*[@id="Awp0GoHome"]'
    home_btn = trying_fnctn(xpath)
    home_btn.click()
    time.sleep(5)
    
dic = {'Item_Id':items, 'State':result_list, 'Version':version_list}

df = pd.DataFrame(dic)

date = date.today()
date = date.strftime('%Y-%m-%d')

df.to_csv('Result_'+date+'.csv')
```
